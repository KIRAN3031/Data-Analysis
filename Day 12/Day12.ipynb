# Transformation Technique

```python
# finding the skewness and the kurtosis values
df['CoapplicantIncome'].skew()
```

```python
df['CoapplicantIncome'].kurtosis()
```

```python
sns.histplot(df['CoapplicantIncome'])
```

```python
from sklearn.preprocessing import PowerTransformer
pt = PowerTransformer(method='yeo-johnson')
```

```python
la_transformed = pt.fit_transform(df[['CoapplicantIncome']])
la_transformed.shape
```

```python
print("Skewness :",pd.Series(la_transformed.reshape(614,)).skew())
print("Kurtosis :",pd.Series(la_transformed.reshape(614,)).kurtosis())
```

```python
sns.histplot(la_transformed)
```
