## Task 2
    1.  Identify and count missing values in each column.

    2. Drop column(s) that are irrelevant for price analysis (example: society or unnamed columns).

    3. Handle missing values in numeric columns such as bath, balcony, etc. (use either dropna() or fillna() depending on your reasoning).

    4. Convert total_sqft to numeric â€” handle values like "2100 - 2850" by converting them to an average or a single number.

    5. Remove duplicate rows.

    6. Reset the DataFrame index after cleaning

```python
# identifing and counting missing values in each column
df.isnull().sum()
```

```python
# droping the society column
df.drop(['society'],axis=1,inplace=True,errors='ignore')
```

```python
# filling the missing values of the bath and balcony
df['bath'].fillna(df['bath'].mean(),inplace = True)
```

```python
df['balcony'].fillna(df['balcony'].median(),inplace = True)
```

```python
# total_sqft cleanup
def convert_sqft(x):
  if isinstance(x,str) and '-' in x:
    a,b = x.split('-')
    return (float(a)+float(b))/2
  try:
    return float(x)
  except:
    return None

df['total_sqft']=df['total_sqft'].apply(convert_sqft)
df.dropna(subset=['total_sqft'],inplace=True)
```

```python
# removing the duplicate rows in the dataset
df.drop_duplicates(inplace=True)
```

## Task 4
    1. Price Distribution:
        Plot the distribution of the price column using a Histogram + KDE curve.

    2. Relationship Between Area and Price:
        Visualize the relationship between total_sqft and price using a Scatter Plot.
        
    3. Effect of Bathrooms on Price:
        Show how bath count affects house prices using a Box Plot.

    4. Top 10 Most Expensive Locations:
        Plot the Top 10 locations with highest average price using a Bar Chart.
        
    5. Correlation Between Numeric Columns:
        Create a Heatmap to visualize correlations among numeric features (price, total_sqft, bath, balcony, etc.).

```python
# importing the neccessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Price Distribution using histogram and KDE curve
plt.figure(figsize=(10,5))
sns.histplot(df['price'],kde=True,bins=40)
plt.title('Price Distribution')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
# Visualize the relationship between total_sqft and price using a Scatter Plot.
plt.figure(figsize=(10,5))
sns.scatterplot(x='total_sqft',y='price',data=df)
plt.xlabel('Total Square Feet')
plt.ylabel('Price')
plt.title('Relationship between Total Square Feet and Price')
plt.grid(True)
plt.show()
```

```python
# Show how bath count affects house prices using a Box Plot.
plt.figure(figsize=(10,5))
sns.boxplot(x='bath',y='price',data=df)
plt.xlabel('Number of Bath')
plt.ylabel('Price')
plt.xticks(rotation=45)
plt.title('Effect of Bath on Price')
plt.grid(True)
plt.show()
```

```python
# Plot the Top 10 locations with highest average price using a Bar Chart.
plt.figure(figsize=(10,5))
sns.barplot(x=df.groupby('location')['price'].mean().sort_values(ascending=False).head(10).index,y=df.groupby('location')['price'].mean().sort_values(ascending=False).head(10).values)
plt.xlabel('Location')
plt.ylabel('Average Price')
plt.title('Top 10 Locations with Highest Average Price')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
# Create a Heatmap to visualize correlations among numeric features (price, total_sqft, bath, balcony, etc.).
plt.figure(figsize=(10,5))
sns.heatmap(df[['price','total_sqft','bath','balcony']].corr(),annot=True,cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

## Task 4 â€” Programming Task: Feature Engineering + Logical Filters


### Task 4A â€” Create New Columns
    Write code to create:
    1. discount_percent = ((actual_price - discount_price) / actual_price) * 100

    2. price_category based on:
        actual_price < 500 â†’ "Budget"
        500 â‰¤ price < 2000 â†’ "Midrange"
        price â‰¥ 2000 â†’ "Premium"

    3. popularity_score = rating * log(rating_count + 1)

```python
'''price_category based on:
    actual_price < 500 â†’ "Budget"
    500 â‰¤ price < 2000 â†’ "Midrange"
    price â‰¥ 2000 â†’ "Premium" '''

df['price_category'] = np.where(df['actual_price'] < 500, 'Budget', np.where(df['actual_price'] < 2000, 'Midrange', 'Premium'))
df
```

```python
# popularity_score = rating * log(rating_count + 1)
df['popularity_score'] = df['rating'] * np.log(df['rating_count'] + 1)
```
